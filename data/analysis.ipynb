```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
image_dir = 'Train'
log = 'eval_log.csv'
```

```python
df = pd.read_csv(log, names=['image', 'loss'])
df = df.sort_values(by='loss', ascending=False)
```

```python
limit = 8
head = df.iloc[:limit]
fig_size = 12
fig = plt.figure(figsize=(fig_size, fig_size))

ctr = 0

for i, row in head.iterrows():
    ctr+=1
    if ctr > limit:
        break

    fig.add_subplot(limit/2,limit/2,ctr)

    image = Image.open(os.path.join(image_dir, row.image))
    image = np.array(image)
    plt.imshow(image[:,:,0], cmap='gray')
    plt.title(f'Loss: {round(row.loss, 2)}')
    plt.axis('off')

plt.show()
```

```python
limit = 8
tail = df[len(df)-limit-1:]
fig_size = 12
fig = plt.figure(figsize=(fig_size, fig_size))

ctr = 0

for i, row in tail.iterrows():
    ctr+=1
    if ctr > limit:
        break

    fig.add_subplot(limit/2,limit/2,ctr)

    image = Image.open(os.path.join(image_dir, row.image))
    image = np.array(image)
    plt.imshow(image[:,:,0], cmap='gray')
    plt.title(f'Loss: {round(row.loss, 2)}')
    plt.axis('off')

plt.show()
```

```python
val_dir = 'Val'
for file in os.listdir(val_dir):
    if '.png' not in file:
        continue
    image = Image.open(os.path.join(val_dir, file))
    plt.figsize(8,8)
    plt.imshow(image)
    plt.title(file)
    plt.show()
```
